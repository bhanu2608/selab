```python
import matplotlib.pyplot as plt
import numpy as np

days = np.arange(1, 31)



temperature = 20 + 2 * days + np.random.normal(0, 5, len(days))

a = 0.5
b = 2
c = 15
iterations = 5

plt.scatter(days, temperature, label='Actual Data')

for i in range(iterations):
    predicted = a * days**2 + b * days + c
    a -= 0.0001 * np.sum((predicted - temperature) * days**2)
    b -= 0.01 * np.sum((predicted - temperature) * days)
    c -= 0.1 * np.sum(predicted - temperature)

    plt.plot(days, predicted, label=f'Iteration {i + 1}')

final_predicted = a * days**2 + b * days + c
plt.plot(days, final_predicted, label='Final Model', linewidth=2)

plt.legend()
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather Modeling Using Quadratic Model')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = {'Category': ['Value 1', 'Value 2', 'Value 3', 'Value 4', 'Value 5'],
        'Value': [30, -20, 10, -5, 25]}
df = pd.DataFrame(data)

cumulative_sum = np.cumsum(df['Value'])

fig, ax = plt.subplots()
ax.bar(df['Category'], df['Value'], color='b', align='center')
ax.set_xlabel('Category')
ax.set_ylabel('Value')
ax.set_title('Waterfall Chart')

prev = 0
for i, val in enumerate(df['Value']):
    if val < 0:
        ax.axhline(y=cumulative_sum[i], color='r', linestyle='--')
        ax.annotate(str(val), xy=(i, cumulative_sum[i]), xytext=(i+0.2, cumulative_sum[i]))
    else:
        ax.axhline(y=cumulative_sum[i], color='g', linestyle='--')
        ax.annotate('+' + str(val), xy=(i, prev), xytext=(i+0.2, prev))
    prev = cumulative_sum[i]

plt.show()
```
